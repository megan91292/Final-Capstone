# Creatin new CSV for root mean square energy

```python
header = 'filename rmse'
```

```python

file = open('rmse.csv', 'w', newline='')
with file:
    writer = csv.writer(file)
    writer.writerow(header)
data_dir = '../data/music/'
for filename in os.listdir(data_dir):
    full_path = f'{data_dir}{filename}'
    trackname = filename.split(".")
    y, sr = librosa.load(full_path, mono=True, duration=30)
    rmse = librosa.feature.rms(y)
    to_append = f'{trackname[0]} {np.mean(rmse)}'
    file = open('rmse.csv', 'a', newline='')
    with file:
        writer = csv.writer(file)
        writer.writerow(to_append.split())
```
